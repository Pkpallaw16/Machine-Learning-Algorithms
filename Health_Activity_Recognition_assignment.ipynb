```python
# suffling data 
from sklearn.utils import shuffle

# test  = shuffle(test)
train_data = shuffle(train_data)
```

```python
# separating data inputs and output lables 
trainData  = train_data.drop('Activity' , axis=1)
trainLabel = train_data.Activity.values
print(trainData.shape)
testData  = test_data.drop('Activity' , axis=1)
testLabel = test_data['Activity'].values
trainData.head()

#trainData = train_data.iloc[:, :-1].values  
#trainLabel = train_data.iloc[:, -1].values  
```

```python
# encoding labels 
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()

# encoding test labels 
encoder.fit(testLabel)
testLabelE = encoder.fit_transform(testLabel)
print(testLabelE.shape)
# encoding train labels 
encoder.fit(trainLabel)
trainLabelE = encoder.fit_transform(trainLabel)
print(trainLabelE.shape)
#encoder = LabelEncoder()
#y =encoder.fit_transform(y)
```

```python
pd.crosstab(train_data.subject, train_data.Activity)
testLabelE = testLabelE.reshape(-1,1)
#trainLabelE = trainLabelE.reshape(-1,1)
print(testLabelE)
```

```python
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn import metrics
svc=SVC() #Default hyperparameters
svc.fit(trainData,trainLabelE)
y_pred=svc.predict(testData)
print('Accuracy Score:')
print(metrics.accuracy_score(testLabelE,y_pred))
print(classification_report(testLabelE,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
classifiers = [
    DecisionTreeClassifier(),
    KNeighborsClassifier(43), # because there are 6 different labels
    SVC(),
    GaussianNB(),
    ]
```

```python
names = []
scores = []
for clf in classifiers:
    clf = clf.fit(trainData,trainLabelE)
    y_pred = clf.predict(testData)
    names.append(clf.__class__.__name__)
    scores.append(metrics.accuracy_score(testLabelE,y_pred))
 

score_df = pd.DataFrame({'Model': names, 'Score': scores}).set_index('Model')
score_df
```

```python
#applying k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier as knn
import numpy as np

knnScoreDistance=np.zeros(51)
knnScoreUniform=np.zeros(51)

for num in range(5,51):
    knnclf = knn(n_neighbors=num, weights='distance')
    knnModel = knnclf.fit(trainData , trainLabelE)
    knnScoreDistance[num]=knnModel.score(testData  , testLabelE )
    print("Testing  set score for KNN_Distance(k=%d): %f" %(num,knnScoreDistance[num]))
    
for num in range(5,51):
    knnclf = knn(n_neighbors=num, weights='uniform')
    knnModel = knnclf.fit(trainData , trainLabelE)
    knnScoreUniform[num]=knnModel.score(testData  , testLabelE )
    print("Testing  set score for KNN_Uniform(k=%d): %f" %(num,knnScoreUniform[num]))
```

```python
import matplotlib.pyplot as plt

x=np.array(range(5,51))

plt.plot(x,knnScoreDistance[5:])
plt.plot(x,knnScoreUniform[5:])
plt.xlabel("No of neighbors (K)")
plt.ylabel("Test Data Mean Accuracy")
plt.legend(['KNN_Distance','KNN_Uniform'])
plt.show()
```
